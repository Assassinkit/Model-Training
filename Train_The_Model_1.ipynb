# **Load, Train and Tune your Data**



# **Load Data**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

```python
# @title AromaticProportion

from matplotlib import pyplot as plt
df['AromaticProportion'].plot(kind='line', figsize=(8, 4), title='AromaticProportion')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title Aromatic Proportion vs Molecular Weight

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['MolWt'], df['AromaticProportion'], c=df['logS'], cmap='viridis', alpha=0.5)
plt.title('Aromatic Proportion vs Molecular Weight')
plt.xlabel('Molecular Weight (MolWt)')
plt.ylabel('Aromatic Proportion')
_ = plt.colorbar(label='Log of Solubility (logS)')
```

# **Data Preparation**

### Data Separation as X and Y

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

## Data splitting

```python
from sklearn.model_selection import train_test_split  ## Importing the Model

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# **Model Building**

## **Linear Regression**

###**Training the model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

###**Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

###**Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred) ##R^2 (coefficient of determination) regression score function.

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## **Random Forest**

###**Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### **Applying the model to make a prediction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
```
